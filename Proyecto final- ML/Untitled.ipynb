```python
import numpy as np
import pandas as pd
from sklearn import linear_model
df_casas = pd.read_csv('casas-pro.csv')
df_departamentos = pd.read_csv('departamentos-pro.csv')
centrosur='centro sur|Centro sur|Centro Sur'
```

```python
df_casas=df_casas.loc[df_casas['titulo'].str.contains(centrosur)]
casas_m = df_casas['mets'].replace(' m²', '', regex=True).astype(int).to_frame() 
casas_p = df_casas['precio'].replace('$|,|MN', '', regex=True).astype(int).to_frame() 
```

```python
model = linear_model.LinearRegression()
model.fit(casas_m ,casas_p )
#model.predict([[400]])
#m=pd.DataFrame([92,110,135,180,300])
#p=pd.DataFrame([2.5,2.8,3.2,4.5,6.2])
#model.fit(m,p)
m2=700
predict=model.predict(pd.DataFrame([m2]))
print('\nPara una casa en Centro Sur de {}m² el precio estimado es de ${} '.
      format(m2,predict[0]))
```

```python
df_departamentos=df_departamentos.loc[df_departamentos['titulo'].str.contains(centrosur)]
departamentos_m = df_departamentos['mets'].replace(' m²', '', regex=True).astype(int).to_frame() 
departamentos_p = df_departamentos['precio'].replace('$|,|MN', '', regex=True).astype(int).to_frame() 
```

```python
model = linear_model.LinearRegression()
model.fit(departamentos_m ,departamentos_p )
m2=100
predict=model.predict(pd.DataFrame([m2]))
print('\nPara un departamento en Centro Sur de {}m² el precio estimado es de ${} '.
      format(m2,predict[0]))
```
