```python
import pandas as pd
from datetime import datetime
df = pd.read_csv('stock-FAANG-mensual.csv')

df = pd.read_csv('stock-empresas10-mensual.csv')
#df = pd.read_csv('stock-empresas10.csv')

df=df.set_index('Empresa')

print('Tamaño del la muestra {}'.format(df['Open'].count()))

#delimitar 2020
df['Date']=pd.to_datetime(df['Date'])
df20=df.loc[(df['Date'] < datetime.strptime('21-01-01', '%y-%m-%d'))]

#delimitar 2021 hasta el 12 de marzo
df['Date']=pd.to_datetime(df['Date'])
df21=df.loc[(df['Date'] >= datetime.strptime('21-01-05', '%y-%m-%d'))
            & (df['Date'] <= datetime.strptime('21-03-12', '%y-%m-%d'))]

df
```

```python
data=df20.loc[df20["Date"]==df20['Date'].min(),'Open'].replace(',', '', regex=True).astype(float).to_frame() 
data=data.rename(columns={"Open": "Open20"})
data['Close20']=df20.loc[df20["Date"]==df20['Date'].max(),'Close*'].replace(',', '', regex=True).astype(float)
data['2020']=data['Close20']*100/data['Open20']
print(df20['Date'].min())
print(data)

print('\nLa Empresa con mejor rendimiento en 2020 fue {} con un crecimiento de {}%'.
      format(data['2020'].idxmax(), round(data['2020'].max(),2)))
```

```python
data['Open21']=df21.loc[df21["Date"]==df21['Date'].min(),'Open'].replace(',', '', regex=True).astype(float)
data['Close21']=df21.loc[df21["Date"]==df21['Date'].max(),'Close*'].replace(',', '', regex=True).astype(float)
data['2021']=data['Close21']*100/data['Open21']
print(df21['Date'].max())
print(df21['Date'].min())
print(data)

print('\nLa Empresa con mejor rendimiento en 2021 fue {} con un crecimiento de {}%'.
      format(data['2021'].idxmax(), round(data['2021'].max(),2)))
```
